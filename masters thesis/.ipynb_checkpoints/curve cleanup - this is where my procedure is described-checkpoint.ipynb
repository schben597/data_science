```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit
from scipy.interpolate import UnivariateSpline, SmoothBivariateSpline

import matplotlib as mpl
mpl.rcParams['svg.fonttype'] = 'none'
mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['font.size'] = 11
```

```python
# Only for here

def identify_dropoff(x, y):
    """
    Check the derivative, going back from the last time point and see where it stops being negative
    Returns x and y for the parts that are dropping off
    """
    
    y_diff = pd.Series(y).diff()
    
    last_val = len(x)

    for i in range(len(x)-1, 0, -1):
        if y_diff[i] > 0:
            
            break
        last_val = i - 1
 
    return x[last_val:], y[last_val:]


######################################################################################################33333

def scanning_diff_std(x, y, lookback, back_offset, lookahead, fwd_offset, nstds):
    
    if lookback != 0:
        if not back_offset < lookback:
            raise ValueError('Back offset too high - no lookback possible')
    if lookahead != 0:
        if not fwd_offset < lookahead:
            raise ValueError('Forward offset too high - no lookahead possible')
    
    y = pd.Series(y)
    
    y_diff = y.diff()
    
    indices = []
    upper_band = []
    lower_band = []
    
    for i in range(lookback, len(y) - lookahead):
        # Expected position based on surrounding average value
        current_y = pd.concat([y[i-lookback:i-back_offset],
                               y[i+1+fwd_offset:i+1+lookahead]]).mean() 
        
        # Exclude current point: take derivatives before and after
        window_diff = pd.concat([y_diff[i-lookback:i-back_offset],
                                 y_diff[i+1+fwd_offset:i+1+lookahead]])
        
        # Stats on surrounding derivatives
        std_diff = window_diff.std()
        

        
        # Bands for outlier detection
        indices.append(i)
        upper_band.append(current_y + nstds * std_diff)
        lower_band.append(current_y - nstds * std_diff)
    
    return indices, upper_band, lower_band

def apply_volatility_bands(x, y, lookback=2, back_offset=1, lookahead=2, fwd_offset=1, nstds=1.5):
    # Apply volatility envelope
    # Returns a pd.Series containing the accepted values
    
    x_scanning, y_scanning_pos, y_scanning_neg = scanning_diff_std(x, y, lookback, back_offset, lookahead, fwd_offset, nstds)
        
    # Get the actual y values for the scanning indices
    check_inside = y[lookback : len(y) - lookahead]
    inside = check_inside[(check_inside < y_scanning_pos) & (check_inside > y_scanning_neg)]

    return inside

def determine_peak_from_tail(y):
    
    diff = y.diff()
    slices = {}
    for i in range(len(diff) - 1, 0, -1):
        # Make slices
        sl = diff[-1:i-1:-1]
        slices[i] = sl
        
    #print(f'Slices: {slices}')
        
    negative_ending_slices = {}
    
    for i, sl in slices.items():

        ends_negative = sl.iloc[-1] < 0

        if ends_negative:
            negative_ending_slices[i] = sl
            
    #print(f'Negative ending slices: {negative_ending_slices}')

            
    negative_changing_slices = []
    
    for i, sl in negative_ending_slices.items():
        if sl.sum() < 0:
            negative_changing_slices.append(sl)
            
    #print(f'Negative changing slices: {negative_changing_slices}')

    
    return negative_changing_slices[-1].index[-1] - 1

def attempt_extrapolate_peak(valid_points):
    
    try:
        peak = determine_peak_from_tail(valid_points)
        
        extrapolated_peak = pd.Series([valid_points[peak]] * (len(x_raw) - peak), index = pd.Index(range(peak, len(x_raw))))
                
        return pd.concat([valid_points[valid_points.index < peak], extrapolated_peak])

    except (IndexError, KeyError) as err:
        #print(f'No peak found. Not extrapolating.')
        return valid_points
    
####################################
# Functions

# Generalised Logistic Function definition
def Gfunc(x, a, b, c, d):
    return d + a*np.exp(-b*np.exp(-c*x))

# Symmetrical derivative definition
def d_fun(x, popt):
    h = 1e-5
    return (Gfunc(x+h, *popt)-Gfunc(x-h, *popt))/(2*h)
```

# Adding Peak Extension

```python
fig, axes = plt.subplots(1,2, figsize=(14,6))
plot_data = data[data['Buffer adjusted']]
plot_data = plot_data[plot_data['Condition'].map(lambda c: c not in ['D4 + 3x peptide', 'D4 + 4x peptide', 'D4 + 5x peptide'])]

plot_data = plot_data[plot_data['exp_id'] == plot_data['exp_id'].unique()[4]]
w = plot_data['Well'].unique()[0]
plot_data = plot_data[plot_data['Well'] == w]

for i in range(2):
    run = i + 1
    
    
    
    x_raw = plot_data[plot_data['Run'] == run].index
    y_raw = plot_data[plot_data['Run'] == run]['Absorbance']
    
    axes[i].scatter(x_raw.values, y_raw.values, color='red', label='Raw Data')
    
    lookback = 2
            
    back_offset = 1
    
    
    lookahead = 2
    
    fwd_offset = 1
    
    nstds = 1.5

    
    
    x_scanning, y_scanning_pos, y_scanning_neg = scanning_diff_std(x_raw, y_raw, lookback, back_offset, lookahead, fwd_offset, nstds)
    
    axes[i].plot(x_scanning,
                 y_scanning_pos,
                 color='red',
                 alpha=0.3,
                 label=f'{lookback} lookback, {back_offset} back offset, {lookahead} lookahead, {fwd_offset} forward offset +/- {nstds} $\sigma$')
    
    axes[i].plot(x_scanning,
                 y_scanning_neg,
                 color='red',
                 alpha=0.3,
                 label=f'{lookback} lookback, {back_offset} back offset, {lookahead} lookahead, {fwd_offset} forward offset +/- {nstds} $\sigma$')
    
    
    
    # Get the actual y values for the scanning indices
    valid_points = apply_volatility_bands(x_raw, y_raw, lookback=2, back_offset=1, lookahead=2, fwd_offset=1, nstds=1.5)       


    
    axes[i].scatter(valid_points.index, valid_points, color='lightblue', label='Valid Points')

    try:
        peak = determine_peak_from_tail(valid_points)
        
        extrapolated_peak = pd.Series([valid_points[peak]] * (len(x_raw) - peak), index = pd.Index(range(peak, len(x_raw))))
        
        axes[i].scatter(extrapolated_peak.index, extrapolated_peak, color='lightgreen', label='Extrapolated Peak')
    
        axes[i].scatter(peak, valid_points[peak], color='green', label='Peak')
        
        adjusted = pd.concat([valid_points[valid_points.index < peak], extrapolated_peak])

    except (IndexError, KeyError) as err:
        print(f'No peak found. Not extrapolating.')
        adjusted = valid_points
            

    
    

    
    
    x = adjusted.index.values
    y = adjusted.values

    try:
        # Calculate data characteristics
        y_min = np.min(y)
        y_max = np.max(y)
        y_range = y_max - y_min
        x_range = np.max(x) - np.min(x)
        
        # Initial parameter guesses
        initial_guess = [
            y_range,        # a: amplitude (data range)
            1.0,           # b: shape parameter (start with 1.0)
            3.0 / x_range, # c: growth rate (reasonable default)
            y_min          # d: lower asymptote (data minimum)
        ]

        # Set parameter bounds
        bounds = (
            [y_range * 0.5,  0.1,  0.01,  y_min - 0.1],  # Lower bounds
            [y_range * 2.0,  10.0, 10.0,  y_min + 0.1]   # Upper bounds
        )

        # Fit the curve
        popt, pcov = curve_fit(Gfunc, x, y, p0=initial_guess, bounds=bounds)
        #popt, pcov = curve_fit(Gfunc, x, y, bounds=bounds)

        # Calculate the derivative and get inflection point
        s_val = 0
        y_spl = UnivariateSpline(x, Gfunc(x, *popt), s=s_val)
        deriv = y_spl.derivative()(x)
        t_max = x[np.argmax(deriv)]

        # Time of reaching the right plateau
        right_plateau = -(Gfunc(t_max, *popt) - d_fun(t_max, popt)*t_max - (popt[3] + popt[0])) / d_fun(t_max, popt)

        if right_plateau <= x[-1]:  # Check if plateau is within experimental timeframe
            the_fit = y_spl(x)

            # Plot the data and fit
            axes[i].plot(x, the_fit, c='blue', label='Logistic fit on adjusted points')
            
            axes[i].axvline(right_plateau, c='blue', ls='--', label='Calculated right plateau')
            
    except (RuntimeError, ValueError) as err:
        print("{err}")
        failed = True
        break
        
        
    

    
plt.legend()
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1,1, figsize=(5.907, 3))
plot_data = data[data['Buffer adjusted']]
plot_data = plot_data[plot_data['Condition'].map(lambda c: c not in ['D4 + 3x peptide', 'D4 + 4x peptide', 'D4 + 5x peptide'])]

plot_data = plot_data[plot_data['exp_id'] == plot_data['exp_id'].unique()[4]]
w = plot_data['Well'].unique()[0]
plot_data = plot_data[plot_data['Well'] == w]

axes = np.array([axes])

for i in range(1):
    run = i + 1
    
    x = plot_data[plot_data['Run'] == run].index.values
    y = plot_data[plot_data['Run'] == run]['Absorbance'].values
    
    try:
        # Calculate data characteristics
        y_min = np.min(y)
        y_max = np.max(y)
        y_range = y_max - y_min
        x_range = np.max(x) - np.min(x)
        
        # Initial parameter guesses
        initial_guess = [
            y_range,        # a: amplitude (data range)
            1.0,           # b: shape parameter (start with 1.0)
            3.0 / x_range, # c: growth rate (reasonable default)
            y_min          # d: lower asymptote (data minimum)
        ]

        # Set parameter bounds
        bounds = (
            [y_range * 0.5,  0.1,  0.01,  y_min - 0.1],  # Lower bounds
            [y_range * 2.0,  10.0, 10.0,  y_min + 0.1]   # Upper bounds
        )

        # Fit the curve
        popt, pcov = curve_fit(Gfunc, x, y, p0=initial_guess, bounds=bounds)
        #popt, pcov = curve_fit(Gfunc, x, y, bounds=bounds)

        # Calculate the derivative and get inflection point
        s_val = 0
        y_spl = UnivariateSpline(x, Gfunc(x, *popt), s=s_val)
        deriv = y_spl.derivative()(x)
        t_max = x[np.argmax(deriv)]

        # Time of reaching the right plateau
        right_plateau = -(Gfunc(t_max, *popt) - d_fun(t_max, popt)*t_max - (popt[3] + popt[0])) / d_fun(t_max, popt)

        if right_plateau <= x[-1]:  # Check if plateau is within experimental timeframe
            the_fit = y_spl(x)

            # Plot the data and fit
            axes[i].plot(x, the_fit, c='red', label='Logistic fit on initial points', alpha=0.5)
            
            axes[i].axvline(right_plateau, c=color_dict['Initial'], ls='--', alpha=0.5, label='Calculated right plateau on original points')
            
    except (RuntimeError, ValueError) as err:
        print("{err}")
        failed = True
        break
    
    
    
    x_raw = plot_data[plot_data['Run'] == run].index
    y_raw = plot_data[plot_data['Run'] == run]['Absorbance']
    
    axes[i].scatter(x_raw.values, y_raw.values, color=color_dict['Initial'], label='Raw Data')
    
    lookback = 2
            
    back_offset = 1
    
    
    lookahead = 2
    
    fwd_offset = 1
    
    nstds = 1.5

    
    
    x_scanning, y_scanning_pos, y_scanning_neg = scanning_diff_std(x_raw, y_raw, lookback, back_offset, lookahead, fwd_offset, nstds)
    
    axes[i].plot(x_scanning,
                 y_scanning_pos,
                 color=color_dict['Envelope'],
                 alpha=0.7,
                 label=f'{lookback} lookback, {back_offset} back offset, {lookahead} lookahead, {fwd_offset} forward offset +/- {nstds} $\sigma$')
    
    axes[i].plot(x_scanning,
                 y_scanning_neg,
                 color=color_dict['Envelope'],
                 alpha=0.7)
    
    
    # Get the actual y values for the scanning indices
    valid_points = apply_volatility_bands(x_raw, y_raw, lookback=2, back_offset=1, lookahead=2, fwd_offset=1, nstds=1.5)       

    axes[i].scatter(valid_points.index, valid_points, color=color_dict['Final'], label='Valid Points')

    

    try:
        peak = determine_peak_from_tail(valid_points)
        
        extrapolated_peak = pd.Series([valid_points[peak]] * (len(x_raw) - peak), index = pd.Index(range(peak, len(x_raw))))
        
        axes[i].scatter(extrapolated_peak.index, extrapolated_peak, color=color_dict['Extrapolated'], label='Extrapolated Peak')
    
        axes[i].scatter(peak, valid_points[peak], color=color_dict['Peak'], label='Peak')
        
        adjusted = pd.concat([valid_points[valid_points.index < peak], extrapolated_peak])

    except (IndexError, KeyError) as err:
        print(f'No peak found. Not extrapolating.')
        adjusted = valid_points
            

    
    
    border = adjusted.index.min()
    
    # Combine unmodified x values and the x values from the adjusted points
    x = pd.concat([pd.Series(x_raw[:border]), pd.Series(adjusted.index)]).values
    
    # Calculate exponential moving average for the unassigned y values based on twice the range
    # Giving much less weight to the first values
    y_ema = pd.DataFrame(y_raw[:border*2]).ewm(alpha=0.9).mean()[:border].values
    
    # Unwrap the arrays and turn into a pd.Series
    y_ema_series = pd.Series([a[0] for a in y_ema])
    
    # Finally combine with the adjusted points
    y = pd.concat([y_ema_series, adjusted]).values
    
    axes[i].scatter(x_raw[:border], y_ema, color=color_dict['Final'])


    try:
        # Calculate data characteristics
        y_min = np.min(y)
        y_max = np.max(y)
        y_range = y_max - y_min
        x_range = np.max(x) - np.min(x)
        
        # Initial parameter guesses
        initial_guess = [
            y_range,        # a: amplitude (data range)
            1.0,           # b: shape parameter (start with 1.0)
            3.0 / x_range, # c: growth rate (reasonable default)
            y_min          # d: lower asymptote (data minimum)
        ]

        # Set parameter bounds
        bounds = (
            [y_range * 0.5,  0.1,  0.01,  y_min - 0.1],  # Lower bounds
            [y_range * 2.0,  10.0, 10.0,  y_min + 0.1]   # Upper bounds
        )

        # Fit the curve
        popt, pcov = curve_fit(Gfunc, x, y, p0=initial_guess, bounds=bounds)
        #popt, pcov = curve_fit(Gfunc, x, y, bounds=bounds)

        # Calculate the derivative and get inflection point
        s_val = 0
        y_spl = UnivariateSpline(x, Gfunc(x, *popt), s=s_val)
        deriv = y_spl.derivative()(x)
        t_max = x[np.argmax(deriv)]

        # Time of reaching the right plateau
        right_plateau = -(Gfunc(t_max, *popt) - d_fun(t_max, popt)*t_max - (popt[3] + popt[0])) / d_fun(t_max, popt)

        if right_plateau <= x[-1]:  # Check if plateau is within experimental timeframe
            the_fit = y_spl(x)

            # Plot the data and fit
            axes[i].plot(x, the_fit, c=color_dict['Final'], label='Logistic fit on adjusted points')
            
            axes[i].axvline(right_plateau, c=color_dict['Final'], ls='--', label='Calculated right plateau on adjusted points')
            
    except (RuntimeError, ValueError) as err:
        print("{err}")
        failed = True
        break
        
        
    

    
#plt.legend()
plt.tight_layout()
plt.ylabel('')
plt.xlabel('')
```
