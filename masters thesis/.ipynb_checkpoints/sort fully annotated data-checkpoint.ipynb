```python
import pandas as pd
import os
```

```python
folder_path = 'runs fully annotated/'

excel_data = {}


for filename in os.listdir(folder_path):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(folder_path, filename)
        sheets = pd.read_excel(file_path, sheet_name='Sheet1', index_col=0)  # Load all sheets as a dict
        excel_data[filename] = sheets
```

```python
# Make sure I can keep runs apart

for k, v in excel_data.items():
    v['exp_id'] = k[:-5]

data = pd.concat([v for k, v in excel_data.items()])

# I need to specify peptide ratio for runs where I just did 1:1

def add1x(cond):
    if cond == 'D4 + peptide':
        return 'D4 + 1x peptide'
    return cond

data['Condition'] = data['Condition'].map(add1x)

data.index.name = 'Time'

data.head()
```

```python
data['NaCl [mM]'] = data['NaCl [mM]'].map(int)

data['Sample Condition'] = data['Condition'] + ' ' + data['NaCl [mM]'].map(str) + ' mM NaCl'
```
