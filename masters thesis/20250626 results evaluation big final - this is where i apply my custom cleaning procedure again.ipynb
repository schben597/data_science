```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import curve_fit
from scipy.interpolate import UnivariateSpline

import matplotlib as mpl
mpl.rcParams['svg.fonttype'] = 'none'
mpl.rcParams['font.family'] = 'Arial'
mpl.rcParams['font.size'] = 11
```

```python
# Generalised Logistic Function definition
def Gfunc(x, a, b, c, d):
    return d + a*np.exp(-b*np.exp(-c*x))

# Symmetrical derivative definition
def d_fun(x, popt):
    h = 1e-5
    return (Gfunc(x+h, *popt)-Gfunc(x-h, *popt))/(2*h)

######################################################################################################33333

def scanning_diff_std(x, y, lookback, back_offset, lookahead, fwd_offset, nstds):
    
    if lookback != 0:
        if not back_offset < lookback:
            raise ValueError('Back offset too high - no lookback possible')
    if lookahead != 0:
        if not fwd_offset < lookahead:
            raise ValueError('Forward offset too high - no lookahead possible')
    
    y = pd.Series(y)
    
    y_diff = y.diff()
    
    indices = []
    upper_band = []
    lower_band = []
    
    for i in range(lookback, len(y) - lookahead):
        # Expected position based on surrounding average value
        current_y = pd.concat([y[i-lookback:i-back_offset],
                               y[i+1+fwd_offset:i+1+lookahead]]).mean() 
        
        # Exclude current point: take derivatives before and after
        window_diff = pd.concat([y_diff[i-lookback:i-back_offset],
                                 y_diff[i+1+fwd_offset:i+1+lookahead]])
        
        # Stats on surrounding derivatives
        std_diff = window_diff.std()
        

        
        # Bands for outlier detection
        indices.append(i)
        upper_band.append(current_y + nstds * std_diff)
        lower_band.append(current_y - nstds * std_diff)
    
    return indices, upper_band, lower_band

def apply_volatility_bands(x, y, lookback=2, back_offset=1, lookahead=2, fwd_offset=1, nstds=1.5):
    # Apply volatility envelope
    # Returns a pd.Series containing the accepted values
    
    x_scanning, y_scanning_pos, y_scanning_neg = scanning_diff_std(x, y, lookback, back_offset, lookahead, fwd_offset, nstds)
        
    # Get the actual y values for the scanning indices
    check_inside = y[lookback : len(y) - lookahead]
    inside = check_inside[(check_inside < y_scanning_pos) & (check_inside > y_scanning_neg)]

    return inside

def determine_peak_from_tail(y):
    
    diff = y.diff()
    slices = {}
    for i in range(len(diff) - 1, 0, -1):
        # Make slices
        sl = diff[-1:i-1:-1]
        slices[i] = sl
        
    #print(f'Slices: {slices}')
        
    negative_ending_slices = {}
    
    for i, sl in slices.items():

        ends_negative = sl.iloc[-1] < 0

        if ends_negative:
            negative_ending_slices[i] = sl
            
    #print(f'Negative ending slices: {negative_ending_slices}')

            
    negative_changing_slices = []
    
    for i, sl in negative_ending_slices.items():
        if sl.sum() < 0:
            negative_changing_slices.append(sl)
            
    #print(f'Negative changing slices: {negative_changing_slices}')

    
    return negative_changing_slices[-1].index[-1] - 1

def attempt_extrapolate_peak(valid_points):
    
    try:
        peak = determine_peak_from_tail(valid_points)
        
        extrapolated_peak = pd.Series([valid_points[peak]] * (len(x_raw) - peak), index = pd.Index(range(peak, len(x_raw))))
                
        return pd.concat([valid_points[valid_points.index < peak], extrapolated_peak])

    except (IndexError, KeyError) as err:
        #print(f'No peak found. Not extrapolating.')
        return valid_points
```

```python
data = pd.read_excel('annotated.xlsx', index_col=0)
```

```python
color_dict = {
    'D4 + 2x peptide': '#457B9D',
    'D4 + 1.5x peptide': '#77ABBD',
    'D4 + 1x peptide': '#A8DADC',
    'D4': '#E63946', 
    'WT': '#1D3557',
    'Buffer': '#232323'
}
```

```python
data['Buffer adjusted'] = data['Buffer adjusted'].map(bool)
```

```python
plot_data = data[data['Buffer adjusted']]

plot_data = plot_data[plot_data['Condition'].map(lambda c: c not in ['Buffer', 'Water'])]
plot_data = plot_data[plot_data['Condition'].map(lambda c: c not in ['D4 + 3x peptide', 'D4 + 4x peptide', 'D4 + 5x peptide'])]

exps = plot_data['exp_id'].unique()
nexps = len(exps)

fig, axes = plt.subplots(nexps, 2, figsize=(5.90666, 20))

# Initialize results dataframe with all the columns we want to keep
original_columns = ['Well', 'Condition', 'Run', 'NaCl [mM]', 'Protein [uM]', 
                   'Well_volume', 'exp_id', 'Sample Condition']
results_columns = original_columns + ['Time condensed [min]', 'Delta Abs Run 1', 'Delta Abs Run 2']
results = pd.DataFrame(columns=results_columns)

for e in range(nexps):  # for each experiment
    exp_data = plot_data[plot_data['exp_id'] == exps[e]]
    
    # Get y-axis limits for this experiment (both runs)
    y_min_exp = exp_data['Absorbance'].min()
    y_max_exp = exp_data['Absorbance'].max()

    conds = exp_data['Condition'].unique()  # all the conditions for this experiment
    
    for cond in conds:  # for each condition in this experiment
        salts = exp_data['NaCl [mM]'].unique()  # all the salts for this condition
        
        for salt in salts:  # now for each of these salt concentrations
            # for each of them i have multiple repetitions
            all_reps = exp_data[(exp_data['Condition'] == cond) & (exp_data['NaCl [mM]'] == salt)]
            reps = all_reps['Well'].unique()
            
            for rep in reps:  # Now we're at the single repetition level
                #col = color_id_dict[' '.join([cond, str(salt)])]
                col = color_dict[cond]
                
                
                # Initialize storage for this repetition
                rep_data = exp_data[(exp_data['Condition'] == cond) & 
                                   (exp_data['NaCl [mM]'] == salt) & 
                                   (exp_data['Well'] == rep)]
                
                # Initialize result row with original data (taking first row's values)
                rep_result = {}
                for column in original_columns:
                    if column == 'Run':
                        continue  # We'll handle Run separately since we need both
                    rep_result[column] = rep_data[column].iloc[0]
                
                # Storage for calculated values
                run1_min = None
                run2_min = None
                time_to_condense_run1 = None
                time_condensed_run1 = None
                time_to_condense_run2 = None
                time_condensed_run2 = None
                delta_abs_run1 = None
                delta_abs_run2 = None
                failed = False
                


                for i in range(2):  # for each run
                    if failed:
                        break
                        
                    run = i + 1  # current run that we're looking at
                    # Establish starting raw values

                    run_data = rep_data[rep_data['Run'] == run]
                    
                    x_raw = run_data.index
                    y_raw = run_data['Absorbance']            
            
                    run_time = x_raw[-1]
            




                    lookback = 2

                    back_offset = 1


                    lookahead = 2

                    fwd_offset = 1

                    nstds = 1.5


                    valid_points = apply_volatility_bands(x_raw, y_raw, lookback, back_offset, lookahead, fwd_offset, nstds)       

                    adjusted = attempt_extrapolate_peak(valid_points)


                    x = adjusted.index.values
                    y = adjusted.values
                    
                    # Cutoff for adjusted values
                    border = adjusted.index.min()

                    # Combine unmodified x values and the x values from the adjusted points
                    x = pd.concat([pd.Series(x_raw[:border]), pd.Series(adjusted.index)]).values

                    # Calculate exponential moving average for the unassigned y values based on twice the range
                    y_ema = pd.DataFrame(y_raw[:border*2]).ewm(alpha=0.9).mean()[:border].values

                    # Unwrap the arrays and turn into a pd.Series
                    y_ema_series = pd.Series([a[0] for a in y_ema])

                    # Finally combine with the adjusted points
                    y = pd.concat([y_ema_series, adjusted]).values
                    

                    try:
                        # Calculate data characteristics
                        y_min = np.min(y)
                        y_max = np.max(y)
                        y_range = y_max - y_min
                        x_range = np.max(x) - np.min(x)

                        # Initial parameter guesses
                        initial_guess = [
                            y_range,        # a: amplitude (data range)
                            1.0,           # b: shape parameter (start with 1.0)
                            3.0 / x_range, # c: growth rate (reasonable default)
                            y_min          # d: lower asymptote (data minimum)
                        ]

                        # Set parameter bounds
                        bounds = (
                            [y_range * 0.5,  0.1,  0.01,  y_min - 0.1],  # Lower bounds
                            [y_range * 2.0,  10.0, 10.0,  y_min + 0.1]   # Upper bounds
                        )

                        # Fit the curve
                        popt, pcov = curve_fit(Gfunc, x, y, p0=initial_guess, bounds=bounds)

                        # Calculate the derivative and get inflection point
                        s_val = 0
                        y_spl = UnivariateSpline(x, Gfunc(x, *popt), s=s_val)
                        deriv = y_spl.derivative()(x)
                        t_max = x[np.argmax(deriv)]

                        # Time of reaching the right plateau
                        right_plateau = -(Gfunc(t_max, *popt) - d_fun(t_max, popt)*t_max - (popt[3] + popt[0])) / d_fun(t_max, popt)

                        if right_plateau <= x[-1]:  # Check if plateau is within experimental timeframe
                            the_fit = y_spl(x)

                            # Plot the data and fit
                            axes[e][i].plot(x, the_fit, c=col)
                            
                            size=0.5

                            if cond == conds[0]:  # Only label the first instance of that condition
                                axes[e][i].scatter(x_raw, y_raw, c=col, label=cond, s=size)
                            else:
                                axes[e][i].scatter(x_raw, y_raw, c=col, s=size)  # No label for other reps

                            if run == 1:
                                axes[e][i].axvline(right_plateau, c=col,  alpha=0.5) # linestyle='--',
                                # Store Run 1 specific values
                                time_to_condense_run1 = right_plateau  # Save Run 1's plateau time
                                time_condensed = run_time - right_plateau
                                delta_abs_run1 = the_fit.max() - the_fit.min()
                                run1_min = the_fit.min()
                                
                                
                            


                            elif run == 2:
                                axes[e][i].axvline(right_plateau, c=col,  alpha=0.5) # linestyle='--',

                                # For Run 2, we only calculate delta absorbance
                                # We do NOT update time_to_condense or time_condensed
                                time_to_condense_run2 = right_plateau
                                delta_abs_run2 = the_fit.max() - the_fit.min()
                                run2_min = the_fit.min()

                                # Optional: you might want to plot Run 2's plateau for comparison
                                # axes[e][i].axvline(right_plateau, c=col, linestyle=':', alpha=0.3)

                        else:
                            failed = True
                            break

                    except (RuntimeError, ValueError) as err:
                        print(f"Fitting failed for {cond}, {salt}, {rep}, Run {run}: {err}")
                        failed = True
                        break
                        
                # After processing both runs, add to results if successful
                if not failed and time_condensed is not None:
                    rep_result['Time condensed [min]'] = time_condensed
                    rep_result['Delta Abs Run 1'] = delta_abs_run1
                    rep_result['Delta Abs Run 2'] = delta_abs_run2
                    rep_result['Time to condense [min]'] = time_to_condense_run1  # Use the saved Run 1 value
                    rep_result['Change in Delta Abs'] = delta_abs_run2 - delta_abs_run1
                    rep_result['Time to condense Run 2 [min]'] = time_to_condense_run2
                    rep_result['Change in time to condense [min]'] = time_to_condense_run2 - time_to_condense_run1
                    rep_result['Change in min abs'] = run2_min - run1_min
                    # Final verification
                    total_expected = x[-1] if x[0] == 0 else x[-1] - x[0]
                    #print(f"Final check for {rep}: {time_to_condense:.2f} + {time_condensed:.2f} = {time_to_condense + time_condensed:.2f}, expected {total_expected:.2f}")

                    # Add to results dataframe
                    results = pd.concat([results, pd.DataFrame([rep_result])], ignore_index=True)
                
                # Set same y-axis limits and titles for both runs of this experiment
                for i in range(2):
                    axes[e][i].set_ylim(y_min_exp, y_max_exp)
                    axes[e][i].set_title(f"{exps[e]} - Run {i+1}")
                    axes[e][i].set_xlabel('')
                    axes[e][i].set_ylabel('')
                    
plt.tight_layout()
#plt.savefig('./figures/run_fit_overview.svg')
```
